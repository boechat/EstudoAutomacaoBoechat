```python
### 1  Abre o Site

from selenium import webdriver
navegador = webdriver.Chrome()
url = ("https://rpachallenge.com/")
##navegador.get vai até a pagina
navegador.get(url)

### 2 Faz o Download

## Puxar o By do Webdriver
from selenium.webdriver.common.by import By
#navegador.find_element(By.XPATH, '/html/body/app-root/div[2]/app-rpa1/div/div[1]/div[6]/a').click()
#Comentado para evitar baixar diversas vezes o mesmo arquivo
```

```python
### 3 Ler a Tabela
import pandas as pd
tabela_excel = pd.read_excel(r'C:\Users\User\Downloads\challenge.xlsx')
#Renomeando a coluna Last Name pois havia um caractere vazio
tabela_excel = tabela_excel.rename(columns={'Last Name ': 'Last Name'})
tabela_excel
```

```python
### 4 Clica em Start
navegador.find_element(By.XPATH, '/html/body/app-root/div[2]/app-rpa1/div/div[1]/div[6]/button').click()
```

```python
### 5 Percorre todos os elementos do Navegador 
### 6  Precisará clicar em Submit 
### 7 Repetir o Processo 10 vezes

titulo = '/label'
input_texto = '/input'
dados = 0
for npaginas in range(10):            #Muda Pagina
    ntabela = npaginas
    #print('Pagina '+str(npaginas))
    for ncampos in range (1,8):
        nav_xpath = '/html/body/app-root/div[2]/app-rpa1/div/div[2]/form/div/div['+ str(ncampos) +']/rpa1-field/div'
        elemento = navegador.find_element(By.XPATH, nav_xpath + titulo)
        texto_elemento = nav_xpath+input_texto
        campo_texto = navegador.find_element(By.XPATH,texto_elemento)
        result = elemento.text
        #coluna_tab=tabela_excel.columns[ncolunas]
        ###################
        for ncolunas in range(7):
            if(result == tabela_excel.columns[ncolunas]):
                print('Numero da Coluna   '+ str(ncolunas))
                print('Result   '+ result)
                dados = tabela_excel.iloc[ntabela][ncolunas]
                print(dados)
                #print(str(ntabela)+str(ncolunas))
                campo_texto.send_keys(str(dados))
                print('===============COLOCOU============')
           
    navegador.find_element(By.XPATH, '/html/body/app-root/div[2]/app-rpa1/div/div[2]/form/input').click()


```

```python
### 8 Precisará coletar a mensagem Final, pegando separadamente a Quantidade_Acertos e Tempo_Gasto
import re
# Usando a biblioteca 're'  (Regular Expressions)
mensagem = navegador.find_element(By.XPATH, '/html/body/app-root/div[2]/app-rpa1/div/div[2]/div[2]')
mensagem = mensagem.text

expressao_reg =re.findall(r'\b\d+\b',mensagem)
#Encontra todas as correspondências do padrão na string e retorna como uma lista.
#  r indica uma string                             ;  \b   indica posição entre um alfanumerico e a borda de um alfanumerico
#  \d+ indica um ou mais dígitos (0-9)             ;  \b   indica posição entre um alfanumerico e a borda de um alfanumerico
expressao_reg
porcentagem_acerto = expressao_reg[0]
quantidade_acertos = expressao_reg[1]
total_acertos = expressao_reg[2]
tempo_gasto = expressao_reg[0]
```

```python
### 9) Irá montar uma mensagem <br>
import datetime
# Importando biblioteca Datetime
horaagora = datetime.datetime.now()
hora=horaagora.strftime('%H:%M:%S')
dia = horaagora.strftime('%d/%m/%Y')

corpoemail = 'O Desafio RPA foi finalizado às '+ hora + ' do dia ' + dia + ' retornou uma porcentagem de acerto de ' + porcentagem_acerto + '% , com ' + quantidade_acertos + ' Acertos em um máximo de ' + total_acertos + ' e a execução teve um tempo gasto de ' + tempo_gasto + 'ms.'
corpoemail
```

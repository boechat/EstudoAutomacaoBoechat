1) Os alunos participaram de uma atividade extracurricular e ganharam pontos extras. Esses pontos extras correspondem a 40% da nota atual de cada um deles.
Com base nisso, crie uma coluna chamada "Pontos_extras" que contenha os pontos extras de cada aluno, ou seja, 40% da nota atual deles.

2) Crie mais uma coluna, chamada "Notas_finais" que possua as notas de cada aluno somada com os pontos extras.

3) Como houve uma pontuação extra, alguns alunos que não tinham sido aprovados antes podem ter sido aprovados agora. Com base nisso, crie uma coluna chamada "Aprovado_final" com os seguintes valores:
  True: caso o aluno esteja aprovado (nota final deve ser maior ou igual a 6);
  False: caso o aluno esteja reprovado (nota final deve ser menor que 6).

4) Faça uma seleção e verifique quais alunos não tinham sido aprovados anteriormente, mas foram aprovados após a soma dos pontos extras.

```python
import pandas as pd
import numpy as np
url = "https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/desafios/alunos.csv"
#1) Os alunos participaram de uma atividade extracurricular e ganharam pontos extras. Esses pontos extras correspondem a 40% da nota atual de cada um deles.
#Com base nisso, crie uma coluna chamada "Pontos_extras" que contenha os pontos extras de cada aluno, ou seja, 40% da nota atual deles.

alunos = pd.read_csv(url)
#lê a url, colocando na variavel alunos
alunos = alunos.fillna(value = 0)
#trata os valores nulos, colocando-os como 0
Pontos_extras = alunos['Notas'] * 0.4
#Atribuindo os pontos extras, criando em variável
alunos['Pontos Extras'] = Pontos_extras
#Criando a Coluna 'Pontos Extras' em 'alunos' e Atribuindo a Variavel Pontos_extras
alunos
```

```python
#2) Crie mais uma coluna, chamada "Notas_finais" que possua as notas de cada aluno somada com os pontos extras.
Notas_finais = alunos['Pontos Extras']+alunos['Notas']
#Somamos os Pontos Extras com Notas Finais e colocamos na nossa tabela, não foi pedido mas tá lá!
alunos['Nota Finais'] = Notas_finais
#Notas_finais
alunos
```

```python
#3) Como houve uma pontuação extra, alguns alunos que não tinham sido aprovados antes podem ter sido aprovados agora. Com base nisso, crie uma coluna chamada "Aprovado_final" com os seguintes valores:
#  True: caso o aluno esteja aprovado (nota final deve ser maior ou igual a 6);
#  False: caso o aluno esteja reprovado (nota final deve ser menor que 6).
alunos['Aprovado'] = alunos['Nota Finais'].apply(lambda x: 'True' if x >= 6.0 else 'False')
#Fiz usando lambda, conforme na estrutura abaixo:
#             df['new column name'] = df['column name'].apply(lambda x: 'value if condition is met' if x condition else 'value if condition is not met')
alunos

#Outro Método; buscando o alfa, que pegaria a query de Notas finais >= 6, depois faria um merge, dando um replace
#alfa = alunos.query('@Notas_finais >= 6.0')
#alunos['Aprovado'] = alfa['Aprovado'].replace([False],True)
#alunos
```

```python
#4) Faça uma seleção e verifique quais alunos não tinham sido aprovados anteriormente, mas foram aprovados após a soma dos pontos extras.

#Pegando as notas dos Alunos Originais, chamando de Notas_antes
Notas_antes = alunos['Notas']
#Agora, faremos referencia ao ponto 2 no qual defiminos Notas_Finais, onde falamos que Notas_finais = alunos['Pontos Extras']+alunos['Notas']

#Agora vamos fazer uma query, pegando a variavel @Notas_antes e @Notas_Finais
antes_reprov = alunos.query('@Notas_antes <= 6.0 and @Notas_finais >= 6.0')
antes_reprov
```

# Junta Arquivos

Esse script tem como objetivo ler um excel (gastos_cartao), selecionar duas abas (CLIENTES e CARTAO) e fazer o tratamento de dados.

Para isso, irá procurar na aba CLIENTES todos aqueles que tem DD 11 (São Paulo).

Depois, irá ver pela ID, comparando o ID_CLIENTE da aba CARTAO para pegar os gastos.

Por fim, irá criar o arquivo CSV resultante.

```python
import pandas as pd
linka = r'C:\Users\User\Documents\RPA_AUTO\EXC_001\Output\NOV23\gastoscartao_NOV23.xlsx'
cam = r"C:\Users\User\Documents\RPA_AUTO\EXC_001\Output\\"
arquivo = r'\gastoscartao_'
nmes = "NOV23"
formato = ".xlsx"

link = cam+nmes+arquivo+nmes+formato
link_do_csv = cam+nmes+'\\'

abacartao = pd.read_excel(link,sheet_name=2)
abaclientes = pd.read_excel(link,sheet_name=1)

# O Cliente demanda que seja enviado Os Gastos de Cartão de todos os Clientes com DDD de São Paulo (11) em separado
```
